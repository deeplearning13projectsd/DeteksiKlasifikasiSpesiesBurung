```python
from PIL import Image
import matplotlib.pyplot as plt


output= 1  #Misalkan nilai output hasil klasifikasi dari model Transfer learning (dengan kelas 1- 10)

def hasilklasifikasi(output):
    if output == 1:
        return "/content/burung a.png"
    elif output == 2:
        return "/content/burung b.png"
    else:
        return "/content/tidak dikenal.png"

klasifikasi = hasilklasifikasi(output)
klasifikasi

image = Image.open(klasifikasi)

plt.imshow(image)
plt.axis('off')
plt.title(f"Gambar: {klasifikasi}")
plt.show()
```
